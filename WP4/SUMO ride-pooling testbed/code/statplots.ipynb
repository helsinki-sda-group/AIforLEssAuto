```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("experimentsstats.csv")
```

```python
df.head()
```

```python
df_pt20 = df[df["percTaxi"]==20]
```

```python
len(df_pt20)
```

```python
df_pt20.set_index("capacity", inplace=True)
```

```python
fig, axs = plt.subplots(figsize=(7, 4))

df_pt20[["avgWaitTime"]].plot(ax=axs, use_index=True, ylabel="Average waiting time (sec)")

fig.savefig("avgWaitTime.jpg", dpi=300)

plt.show()
```

```python
fig, axs = plt.subplots(figsize=(7, 4))

df_pt20[["avgDetourDistancePerc", "avgDetourTimePerc", "taxiPrivateCoeff"]].plot(ax=axs, use_index=True, ylabel="Percent")

fig.savefig("passMeasures.jpg", dpi=300)

plt.show()
```

```python
fig, axs = plt.subplots(figsize=(7, 4))

df_pt20[["taxiAvgOccRate", "taxiAvgIdleDistanceRatio", "taxiAvgIdleTimeRatio"]].plot(ax=axs, use_index=True, ylabel="Ratio")

fig.savefig("taxiMeasures.jpg", dpi=300)

plt.show()
```

```python
fig, axs = plt.subplots(figsize=(7, 4))

df_pt20[["advCO", "advCO2", "advHC"]].plot(ax=axs, use_index=True, ylabel="Advantage (%)")

fig.savefig("emissionMeasures.jpg", dpi=300)

plt.show()
```
