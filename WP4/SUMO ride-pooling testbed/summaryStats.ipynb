```python
import pandas as pd
import os
import re
```

```python
experimentsFolder="saved experiments"
currentDir = os.getcwd()
experimentsFolder = currentDir + "\\" + experimentsFolder

expDirs = [ f.path for f in os.scandir(experimentsFolder) if f.is_dir() ]
```

```python
def getint(s):
    digits = re.findall(r'\d+', s)
    if len(digits)!=1:
        print("Wrong format of int string, found " + str(len(digits)) + " numbers")
    else:
        return int(digits[0])
```

```python
def getfloat(s):
    digits = re.findall(r'[-]?\d+.\d+', s)
    if len(digits)!=1:
        print("Wrong format of float string, found " + str(len(digits)) + " numbers")
    else:
        return float(digits[0])
```

```python
data = []

for exp in expDirs:
    statsFolder = exp + "\\" + "output\\stats"
    summaryFile = open(statsFolder + "\\summary.txt",'r')
    contents = summaryFile.readlines()
    trips = getint(contents[2])
    percpass = getint(contents[3])
    taxiPassCount = getint(contents[4])
    percTaxi = getint(contents[5])
    taxiCount = getint(contents[6])
    taxiSharedCount = getint(contents[60])
    taxiNonsharedCount = getint(contents[61])
    routingAlgorithm = contents[9].split(" ")[2].strip()
    capacity = getint(contents[10])
    finishedTrips = getint(contents[15])
    unfinishedTrips = getint(contents[18])
    unstartedTrips = getint(contents[19])
    avgWaitTime = getfloat(contents[23])
    avgDetourTimePerc = getfloat(contents[25])
    avgDetourDistancePerc = getfloat(contents[27])
    taxiPrivateCoeff = getfloat(contents[28])
    taxiAvgCustomers = getfloat(contents[65])
    taxiAvgOccRate = getfloat(contents[66])
    taxiAvgIdleDistanceRatio = getfloat(contents[67])
    taxiAvgIdleTimeRatio = getfloat(contents[68])
    advCO = getfloat(contents[108])
    advCO2 = getfloat(contents[109])
    advHC = getfloat(contents[110])

    row = [trips, percpass, taxiPassCount, percTaxi, taxiCount, taxiSharedCount, taxiNonsharedCount, routingAlgorithm, capacity, finishedTrips, unfinishedTrips, unstartedTrips, \
       avgWaitTime, avgDetourTimePerc, avgDetourDistancePerc, taxiPrivateCoeff, taxiAvgCustomers, taxiAvgOccRate, taxiAvgIdleDistanceRatio, taxiAvgIdleTimeRatio, \
        advCO, advCO2, advHC]
    data.append(row)

    
```

```python
df = pd.DataFrame(data, columns=['trips', 'percpass', 'taxiPassCount', 'percTaxi', 'taxiCount', 'taxiSharedCount', 'taxiNonsharedCount', \
                                 'routingAlgorithm', 'capacity', 'finishedTrips', 'unfinishedTrips', 'unstartedTrips', \
       'avgWaitTime', 'avgDetourTimePerc', 'avgDetourDistancePerc', 'taxiPrivateCoeff', 'taxiAvgCustomers', 'taxiAvgOccRate', 'taxiAvgIdleDistanceRatio',\
          'taxiAvgIdleTimeRatio', 'advCO', 'advCO2', 'advHC'])
```
